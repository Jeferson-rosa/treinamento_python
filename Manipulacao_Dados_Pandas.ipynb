# <font color=green>1 - Importação do pacote Pandas e criação de um Datframe</font>
***

```python
import pandas as pd
import numpy as np
```

```python
# criando o DataFrame com dados aleatórios
np.random.seed(42)
uf = ['SP', 'RJ', 'MG', 'BA', 'SC', 'RS']
sexo = ['M', 'F']
raca = ['Branca', 'Parda', 'Preta', 'Amarela', 'Indígena']
idade = np.random.randint(18, 70, size=30)
anos_de_estudo = np.random.randint(1, 20, size=30)
renda = np.random.randint(1000, 10000, size=30)
altura = np.round(np.random.normal(loc=1.70, scale=0.10, size=30), 2)
dados = pd.DataFrame({'UF': np.random.choice(uf, size=30),
                   'Sexo': np.random.choice(sexo, size=30),
                   'Idade': idade,
                   'Raça': np.random.choice(raca, size=30),
                   'Anos de Estudo': anos_de_estudo,
                   'Renda': renda,
                   'Altura': altura})

# exibindo o DataFrame
print(dados)
```

##<font color=green>2.3 - Selecionando subconjuntos de linhas e colunas com loc e iloc:</font>
***
>Quando usar loc ou iloc?
>
>Basicamente os dois métodos servem para resgatar dados, mas possuem características distintas na hora que vamos utilizá-los.

###<font color=green>2.3.1 - loc:</font>
***
>Este método é primariamente baseado nas labels da colunas, porém podemos utilizar com um array booleano também.

```python
# selecionando uma única célula com loc
print(dados.loc[0, 'UF'])
```

```python
# selecionando um subconjunto de células com loc
print(dados.loc[1:3, ['UF', 'Sexo']])
```

###<font color=green>2.3.2 - iloc:</font>
***
>O indexador iloc é um tanto mais simples, ele seleciona por números inteiros das linhas, arrays ou por slice.

```python
# selecionando uma única célula com iloc
print(dados.iloc[0, 1])
```

```python
# selecionando um subconjunto de células com iloc
print(dados.iloc[1:3, 0:2])
```

```python
classes = [18, 31, 44, 57, 70]
labels = ['18-31', '31-44', '44-57', '57-70']
```

```python
frequencia = pd.value_counts(
    pd.cut(x = dados.Idade,
               bins = classes,
               labels = labels,
               include_lowest = True)
)
frequencia
```

```python
percentual = pd.value_counts(
    pd.cut(x = dados.Idade,
               bins = classes,
               labels = labels,
               include_lowest = True),
    normalize = True
)
percentual
```

```python
  dist_freq_personalizadas = pd.DataFrame(
      {'Frequência': frequencia, 
       'Porcentagem (%)': round(percentual*100, 2)}
  )
  dist_freq_personalizadas
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Criação da área de plotagem do gráfico e o tamanho
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))

# Dados para compor o gráfico
recipe = labels

# Definindo a quantidade de cada dado
dados = frequencia

# Como os valores vão aparecer no gáfico.
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}\n({:.1f}%)".format(absolute, pct)

# Criação do gráfico e inserção dos dados formatados
wedges, texts, autotexts = ax.pie(dados, autopct=lambda pct: func(pct, dados),
                                  textprops=dict(color="w"))

# Configuração da legenda
ax.legend(wedges, recipe,
          title="Idades",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

# Tamanho do texto dentro do gráfico, e fonte em negrito
plt.setp(autotexts, size=8, weight="bold")

# Título do gráfico
ax.set_title("Entrevistados por faixa de idade:")

# Mostrando o gráfico
plt.show()
```
